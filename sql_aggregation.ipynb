```python
%%sql
CREATE TABLE Sales (
    sale_id INT PRIMARY KEY,
    customer_name VARCHAR(50),
    city VARCHAR(50),
    product VARCHAR(50),
    category VARCHAR(50),
    amount DECIMAL(10,2),
    sale_date DATE
);
```

```python
%%sql
SELECT city, SUM(amount) AS total_sales
FROM Sales
GROUP BY city;
```

```python
%%sql
SELECT category, COUNT(*) AS total_orders
FROM Sales
GROUP BY category;
```

```python
%%sql
SELECT city, SUM(amount) AS total_sales
FROM Sales
GROUP BY  city
HAVING SUM(amount) > 65000;
```

```python
%%sql
SELECT * FROM Sales
WHERE city in ('Salem','Madurai');
```

```python
%%sql
SELECT * FROM Sales
WHERE category not in ('Electronics');
```

```python
%%sql
SELECT * FROM Sales
WHERE city = 'Chennai'
AND amount > 10000;
```

```python
%%sql
SELECT * FROM Sales
WHERE product = 'Laptop' 
OR product = 'Mobile';
```
