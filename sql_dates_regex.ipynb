```python
%%sql
CREATE TABLE events (
    event_id INT AUTO_INCREMENT PRIMARY KEY,
    event_name VARCHAR(100),
    event_date DATETIME,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
```

```python
%%sql
INSERT INTO events (event_name, event_date)
VALUES
    ('Team Meeting', '2026-02-16 09:30:00'),
    ('Conference', '2026-02-28 13:30:00'),
    ('Online Webninar', '2026-03-02 16:00:00'),
    ('Workshop', '2026-03-16 10:30:00');
```

```python
%%sql
SELECT * FROM events;
```

```python
%%sql
SELECT event_id,event_name,DAY(event_date) AS event_day,MONTH(event_date) AS event_month,YEAR(event_date) AS event_year FROM events;
```

```python
%%sql
SELECT event_id,event_name,event_date,event_date + INTERVAL 30 DAY AS next_date FROM events;
```

```python
%%sql
SELECT event_id,event_name,DATE_FORMAT(event_date, '%w, %M %e, %Y') AS formtted_date FROM events;
```

```python
%%sql
SELECT event_id,event_name,DATE_FORMAT(event_date, '%h:%i %p') AS formtted_time FROM events;
```

```python
%%sql
CREATE TABLE regex_samples (
    id INT AUTO_INCREMENT PRIMARY KEY,
    object VARCHAR(100)
);
```
