```python
%%sql
DROP TABLE IF EXISTS CustomerDetails;
CREATE TABLE CustomerDetails (
    id INT PRIMARY KEY,
    customer_name VARCHAR(100),
    email VARCHAR(100),
    city VARCHAR(50),
    phone_number VARCHAR(20)
);
```

```python
%%sql

SELECT
length (customer_name) as name_length,
upper(city) as city_name,
lower(customer_name) as cust_name,
concat(customer_name, '-',city,'-','India') as Name,
SUBSTRING(customer_name,1,5) as name_prefix,
TRIM(city) as trimmed_city,
LPAD(customer_name,15,'*') as left_padded_name,
REPLACE(customer_name, ' ','_') as updated_name,
INSTR(customer_name, 'a') as position_of_a,
right(customer_name,5) AS firsgt_5_chars,
REVERSE(city) as reversed_city,
FORMAT(987653210,2) as formatted_number

from CustomerDetails;
```
